**RGB Color images using matplotlib and open cv** \
In the images produced, for each channel, the very intense white colors for each channel indicates the contribution of each color to the colored image. For example, the red channel's black and white appears very overexposed which means there is a high contribution of red in the colored image

```python
red_channel = colored_image[:,:,0]
green_channel = colored_image[:,:,1]
blue_channel = colored_image [:,:,2]

plt.figure(1)
plt.subplot(231) #2 rows 3 columns
plt.imshow(colored_image)
plt.xticks([])
plt.yticks([])

plt.subplot(232) #2 rows 3 columns
plt.imshow(colored_image)
plt.xticks([])
plt.yticks([])

plt.subplot(233) #2 rows 3 columns
plt.imshow(colored_image)
plt.xticks([])
plt.yticks([])

plt.subplot(234) #2 rows 3 columns
plt.imshow(red_channel, cmap = 'gray')
plt.xticks([])
plt.yticks([])
plt.title('red channel')

plt.subplot(235) #2 rows 3 columns
plt.imshow(green_channel, cmap = 'gray')
plt.xticks([])
plt.yticks([])
plt.title('green channel')

plt.subplot(236) #2 rows 3 columns
plt.imshow(blue_channel, cmap= 'gray')
plt.xticks([])
plt.yticks([])
plt.title('blue channel')

plt.show()
```

#RGB to HSV

```python
def f_rgb_to_hsv(r,g,b,scaleFactor):
  r, g, b = r / 255.0, g / 255.0, b / 255.0 #normalization between 0 to 1
  cmax = max(r,g,b)
  cmin = min(r,g,b)
  
  V = cmax * scaleFactor
  
  if cmax == 0:
    S = 0
  else:
    S = (cmax - cmin) / cmax
    S = S * scaleFactor
    
  if cmax == cmin:
    H = 0

  elif cmax == r:
    H = ((g - b) / (cmax-cmin)) * 60
    H = (H % 360)
  elif cmax == g:
    H = (2 + (b - r) / (cmax - cmin)) * 60
    H = (H % 360)

  elif cmax == b:
    H = (4 + (r - g) / (max - min)) * 60
    H = H % 360
  
  if H < 0:
    H = H + 360

  return H,S,V
```

```python
print(f_rgb_to_hsv(100,200,50,100))
```
