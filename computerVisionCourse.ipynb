Grey Scale image manipulation

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = plt.imread(r'albert-einstein_gray.jpg')
plt.imshow(image, cmap='gray')
```

```python
image2 = image.copy()
image2[350:425, 325:400] = 255
image2[350:425, 425:500] = 255
plt.imshow(image2, cmap='gray')
```
