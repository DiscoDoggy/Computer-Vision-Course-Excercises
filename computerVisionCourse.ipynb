**Image Processing Chapter 1: Grey Scale Images** \
Grey Scale image manipulation

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from google.colab.patches import cv2_imshow
```

```python
image = plt.imread(r'albert-einstein_gray.jpg')
plt.imshow(image, cmap='gray')
```

**Image Processing Chapter 1: Excercise 2** \
Blotting Einstein's eyes out with white squares

```python
image2 = image.copy()
image2[350:425, 325:400] = 255
image2[350:425, 425:500] = 255
plt.imshow(image2, cmap='gray')
```

**Greyscale images using OpenCV**



```python
image = cv2.imread(r'albert-einstein_gray.jpg', cv2.IMREAD_GRAYSCALE) #read image as grayscale
cv2_imshow(image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
